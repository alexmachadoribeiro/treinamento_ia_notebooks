Note que conforme o valor de $x$ aumenta em magnitude, o valor de $f(x)$ (representado pelo eixo $y$) também aumenta - mas aumenta em "velocidades" diferentes dependendo do valor de $x$. ##### Gradiente
Diferente do caso univariado, a função agora tem várias "direções" possíveis para calcularmos as taxas de variação. Porém, com as derivadas parciais, podemos construir o *gradiente*, denotado pelo símbolo *nabla* ($\nabla$) da seguinte forma:
<br>
$$\Huge
\begin{aligned}
 \nabla f(x, y) = \begin{bmatrix}
                   \frac{\delta f}{\delta x} \\
                   \frac{\delta f}{\delta y}
                  \end{bmatrix}
\end{aligned}
$$
<br>
O gradiente de uma função em um ponto é um vetor cuja direção representa a direção de máximo aumento da função e e a magnitude deste vetor quantifica esse aumento. Note que no caso que vimos de $f(x, y) = x^2 + y^2$, a função possui um valor mínimo global e ele está no ponto $x=0, y=0$. Suponha que estejamos em um outro ponto $(x, y)$ qualquer da função e calculamos o gradiente neste ponto, obtendo a direção de máximo aumento da função. Podemos dar "um passo" na direção contrária do gradiente, obtendo a máxima diminuição da função. Se dermos vários passos sempre recalculando o gradiente e andando na direção contrária, eventualmente chegaremos no ponto mínimo da função. Esta é uma noção fundamental para o treinamento de redes neurais e é o que dá nome à técnica de "descida de gradiente", ou _gradient descent_.

##### Espaços vetoriais
Um espaço vetorial (ou espaço linear) é um conjunto cujos elementos (vetores) podem ser adicionados entre si e multiplicados por números (escalares), e o resultado sempre é outro elemento do conjunto. Essas operações (adição e multiplicação por escalar) precisam satisfazer algumas regras para serem definidas para todo o espaço vetorial, como existência de vetor nulo (e escalar zero). Um espaço vetorial pode ser visto como um "número multidimensional", onde cada dimensão representa uma direção e uma magnitude. Para as finalidades de machine learning, os conjuntos de dados constituem um espaço vetorial onde cada variável é uma dimensão. Uma amostra então é um vetor neste espaço vetorial, com o valor de cada variável correspondendo à uma coordenada do vetor.

##### Base
A base de um espaço vetorial pode ser vista como o "sistema de coordenadas" que estamos utilizando. Podemos representar qualquer ponto no espaço (vetor) como uma combinação linear dos vetores da base. Isto é, no caso bidimensionall

<br>
$$\Huge
\begin{aligned}
 \vec{v} = a\cdot\vec{x} + b\cdot\vec{y}
\end{aligned}
$$
<br>

onde $x$ e $y$ são os vetores da **base**. Um mesmo espaço vetorial pode ser representado com infinitas bases diferentes.

##### Dependência linear
A base de um espaço vetorial sempre terá $n$ vetores **linearmente independentes**, onde $n$ é a dimensão do espaço. Um vetor $\vec{v}$ é dito linearmente independente de outro conjunto de vetores $\{\vec{w_1}, \vec{w_2}\}$ se não existirem escalares $a$ e $b$ tais que

<br>
$$\Huge
\begin{aligned}
 \vec{v} = a\cdot\vec{w_1} + b\cdot\vec{w_2}
\end{aligned}
$$
<br>

Caso esta equação tenha solução, dizemos que o vetor é **linearmente dependente** do conjunto. Por exemplo, considere os vetores $\vec{v}=\begin{bmatrix}1\\1\end{bmatrix}$ e $\vec{w}=\begin{bmatrix}3\\3\end{bmatrix}$. Note que $\vec{w}$ é linearmente dependente de $\vec{v}$, pois $\vec{w}=3\cdot\vec{v}$. Reciprocamente, $\vec{v}$ é linearmente dependente de $\vec{w}$.

##### Sistemas lineares
Um sistema linear é um conjunto de equações lineares multivariadas. Os sistemas lineares podem ter zero, uma ou infinitas soluções.

<br>
$$\LARGE
\begin{aligned}
\begin{cases}
  x+y+z = 1\\
  3x-y = \frac{5}{2}\\
  5x+z = 5
  \end{cases}
\end{aligned}
$$
<br>

Podemos preencher os coeficientes que faltam com zeros e explicitar os coeficientes 1:

<br>
$$\LARGE
\begin{aligned}
\begin{cases}
  1x+1y+1z = 1\\
  3x-1y+0z = \frac{5}{2}\\
  5x+0y+1z = 5
  \end{cases}
\end{aligned}
$$
<br>

O ponto $\vec{v} = \begin{bmatrix}x\\y\\z\end{bmatrix}$ pode ser visto como um vetor em um espaço tridimensional. Neste contexto, as equações podem ser vistas como uma matriz $A$ de seus coeficientes:
<br>
$$\Large
\begin{aligned}
A = \begin{bmatrix}1&1&1\\3&-1&0\\5&0&1\end{bmatrix}
\end{aligned}$$

E o sistema linear pode então ser representado como:

<br>
$$\Large
\begin{aligned}
\begin{bmatrix}1&1&1\\3&-1&0\\5&0&1\end{bmatrix} \cdot \begin{bmatrix}x\\y\\z\end{bmatrix}
= \begin{bmatrix}1\\\frac{5}{2}\\5\end{bmatrix}
\end{aligned}$$

Ou

$$\Large
\begin{aligned}
A \cdot \vec{v} = \begin{bmatrix}1\\\frac{5}{2}\\5\end{bmatrix}
\end{aligned}$$

##### Transformações lineares
Matrizes (ou sistemas lineares) podem ser vistos como equações multidimensionais e, nesse contexto, são chamadas de transformações lineares. Estas funções têm como entrada um vetor e como saída outro vetor. Estas transformações satisfazem algumas propriedades - dentre elas, sendo $f$ uma transformação linear e $a$ um escalar:

<br>
$$\Large
\begin{gathered}
f(\vec{v_1} + \vec{v_2}) = f(\vec{v_1}) + f(\vec{v_2}) \\
\text{e}\\
f(a \cdot \vec{v}) = a\cdot f(\vec{v})
\end{gathered}$$

##### Produto interno
Podemos definir também uma operação chamada de _produto interno_. Esta operação recebe dois vetores e retorna um escalar, e é comumente denotada pelo símbolo $\cdot$. Podemos definir este produto de diversas formas, mas usaremos a definição canônica "tradicional": o produto interno entre dois vetores é a soma do produto coordenada a coordenada, isto é:

$$\Large
{\vec{a}\cdot\vec{b}=\sum _{i=1}^{n}a_{i}b_{i}=a_{1}b_{1}+a_{2}b_{2}+\cdots +a_{n}b_{n}}$$

Uma das propriedades desta operação é a possibilidade de calcular _distâncias_. Distâncias são funções matemáticas com algumas propriedades especiais, e também podem ser definidas de diversas formas. Com esta definição em particular, a distância no nosso espaço é a chamada _distância euclidiana_, onde tomamos a raíz quadrada da soma dos quadrados das coordenadas. Desta distância surge o famoso teorema de Pitágoras:



A distância euclidiana com essa definição é a raíz do produto interno de um vetor consigo mesmo, isto é:

$$\Large
{\sqrt{\vec{a}\cdot\vec{a}}=\sqrt{\sum _{i=1}^{n}a_{i}^2}=\sqrt{a_{1}^2+a_{2}^2+\cdots +a_{n}^2}}$$

### Natureza da Informação
Como o objetivo de modelos de machine learning é extrair informação de dados, isto é, estimar funções, é interessante podermos mensurar e quantificar a informação. Felizmente, Claude Shannon fez um ótimo trabalho formalizando matematicamente esse conceito em 1948 em um artigo chamado "A Mathematical Theory of Communication". Este artigo pode ser visto como o nascimento formal do _bit_, a unidade fundamental de informação.

Além do bit, Shannon também cunhou o conceito de _entropia de informação_, uma medida fundamentalmente estatística que tenta quantificar a "informação" contida em uma variável aleatória e pode ser calculada da seguinte forma:

<br>
$$\large
{\displaystyle \mathrm {H} (X):=-\sum _{x\in {\mathcal {X}}}p(x)\log p(x)}$$

Essa entropia de informação é uma medida crucial para o funcionamento de algumas técnicas de machine learning e dá origem a diversas outras estatísticas úteis para a área, como _informação mútua_ (https://en.wikipedia.org/wiki/Mutual_information), _entropia condicional_ (https://en.wikipedia.org/wiki/Conditional_entropy) e _entropia cruzada_ (https://en.wikipedia.org/wiki/Cross-entropy).

##### Variáveis aleatórias
Uma _variável aleatória_ é meramente uma formalização matemática de um número (ou objeto) cujo valor depende de eventos aleatórios. As variáveis aleatórias que só assumem uma quantidade finita de valores - como jogar uma moeda, cujo resultado só pode ser cara ou coroa - são chamadas de _variáveis aleatórias discretas_, enquanto as que assumem um valor real qualquer são chamadas de _variáveis aleatórias contínuas_. Podemos fazer "medições" em uma variável aleatória (amostrar) a fim de descobrir seus possíveis valores, além de usá-las como variáveis em funções.

##### Distribuições de probabilidade
Se amostrarmos uma variável algumas vezes, podemos começar a estabelecer uma noção de seu comportamento. A função que descreve a frequência de cada valor possível em uma amostragem é chamada de _distribuição de probabilidade_. A distribuição mais comum e prevalente é a _distribuição normal_, ou _gaussiana_, ou ainda "curva do sino". No caso de jogar uma moeda honesta, nossa distribuição de probabilidade é dita _uniforme_ e pode ser representada como $p(cara) = 0.5$ e $p(coroa) = 1-p$.
<br>


<center><i>Exemplos de distribuições normais</i></center>

<br>
Agora, imagine a reta dos números reais e suponha que estamos no ponto $x=0$. Vamos jogar a moeda honesta e "dar um passo para a direita" caso saia cara, e um passo para a esquerda caso saia coroa. A frequência de cada ponto após $n$ passos se aproxima de uma distribuição normal conforme $n$ tende ao infinito. Vamos visualizar este fenômeno!

```python
# Vamos utilizar o módulo random encapsulado no numpy
# e o pandas para plottar a distribuição
import numpy as np

# Inicializando uma lista vazia e definindo de iterações
pos_final = []
n = 2000
m = 100
for i in range(1, n):
    pos = 0
    for j in range(1, m):
        pos = pos + np.random.choice([-1, 1])
    pos_final = pos_final + [pos]
    
# Plottando um histograma utilizando o pandas
import pandas as pd
df = pd.DataFrame(pos_final, columns=['posicao_final'])
df['posicao_final'].hist(bins=df['posicao_final'].nunique())
```

##### Médias e medianas

Dada uma sequência de amostras de uma variável, existem duas estatísticas especiais extremamente úteis: a média e a mediana. A média é definida como a soma das amostras dividida pela quantidade de amostras:

$$\large
{\displaystyle {\bar {x}}={\frac {1}{n}}\left(\sum _{i=1}^{n}{x_{i}}\right)={\frac {x_{1}+x_{2}+\cdots +x_{n}}{n}}}$$
<br>
A mediana é definida como a "amostra do meio" se ordenarmos as amostras pelo valor - isto é, metade das amostras são maiores que a mediana, e a outra metade é menor.

##### Quantis
A "amostra do meio", ou mediana, representa um "corte" no meio da lista de amostras, isto é, em 50%. Também podemos dizer que é o _percentil 50_ - e definir, por exemplo, o _percentil 25_ como um corte em 25% da lista, tendo 25% das amostras abaixo e 75% acima. Essa noção pode ser generalizada para qualquer particionamento das amostras e o nome geral do conceito é _quantil_. É comum vermos aplicações em _decis_, dividindo o dataset de 10 em 10%, _vintis_, de 5 em 5%, e _quartis_, de 25 em 25%.

```python
# Calculando a média e a mediana do código anterior utilizando o pandas
print('Média:', df['posicao_final'].mean())
print('Mediana (50%):', df['posicao_final'].median())

# Calculando um quantil arbitrário
# Primeiro quartil - 25%
print('25%:', df['posicao_final'].quantile(0.25))
print('75%:', df['posicao_final'].quantile(0.75))
print('42%:', df['posicao_final'].quantile(0.42))
print('19.319%:', df['posicao_final'].quantile(0.19319))
print('95%:', df['posicao_final'].quantile(0.95))
```

##### Variância e Desvio Padrão
Além da média e mediana, existem outras estatísticas fundamentais para descrever as distribuições: a variância e o desvio padrão. Um pode ser definido com o outro, sendo o desvio padrão a raíz quadrada da variância.

A variância descreve, de forma geral, o quão dispersas (i.e. longe da média) estão as amostras. No caso da distribuição normal, tudo que precisamos para parametrizá-la é de sua média e de sua variância: a média nos diz onde ela está centrada - no exercício anterior, 0 - e a variância nos diz o quão "esticada" ela está, isto é, o quão longe da média estão os pontos.

```python
# Calculando variância e desvio padrão com o pandas
print('Variância:', df['posicao_final'].var())
print('Desvio padrão:', df['posicao_final'].std())
```

Uma distinção importante é a unidade de medida. A variância é medida com unidades quadráticas: se estivermos amostrando o peso de pessoas em $kg$, por exemplo, a variância será em $kg^2$. Como é uma medida um pouco mais difícil de interpretar, usamos o desvio padrão, que tem a mesma unidade de medida da variável. Nesse caso, observamos que desviamos em média $9.95$ unidades do centro zero (média).

Note que a média é exatamente o ponto onde iniciamos. Vamos executar novamente a simulação iniciando na posição $5$:

```python
# Executando mais iterações
n = 50000
m = 100
pos_final = []
for i in range(1, n):
    pos = 15
    for j in range(1, m):
        pos = pos + np.random.choice([-1, 1])
    pos_final = pos_final + [pos]
df = pd.DataFrame(pos_final, columns=['posicao_final'])
df['posicao_final'].hist(bins=df['posicao_final'].nunique())
```

##### Gaussianas
Vamos estimar a distribuição que aferimos fazendo a simulação. Para parametrizarmos uma distribuição normal, precisaremos da média e desvio padrão (ou variância) desta - neste caso, 0 e 9.95. Sua forma geral é dada pela equação

<br>
$$\Large
{\displaystyle f(x)={\frac {1}{\sigma {\sqrt {2\pi }}}}e^{-{\frac {1}{2}}\left({\frac {x-\mu }{\sigma }}\right)^{2}}}$$

Vamos calculá-la para vários pontos e sobrepor ao histograma observado.

```python
# Plotando os dados observados com o matplotlib
# e sobrepondo com uma gaussiana calculada
import matplotlib.pyplot as plt

# Definindo função gaussiana com os parâmetros calculados - média 0 e desvio padrão 9.95
# O parâmetro h representa a contagem máxima e escalona a distribuição estimada verticalmente
def gauss(x, dp=9.95, h=8000, media=0):
    from math import exp, pow
    variancia = pow(dp, 2)
    return h * exp(-pow(x-media, 2)/(2*variancia))

# Definindo contagem máxima para escalonamento da gaussiana
max_val = df['posicao_final'].value_counts().max()

# Criando lista de pontos do eixo x e f(x)
x = np.arange(-35, 45, 0.01)
yg = [gauss(xi, media=15, h=max_val) for xi in x]

plt.plot(x,yg)
plt.hist(df['posicao_final'], bins=df['posicao_final'].nunique())
plt.show()
```

##### Normalização e padronização
Muitas das técnicas de machine learning assumem distribuições de "tamanhos" parecidos - isto é, com médias similares e desvios padrões similares. A maioria das distribuições encontradas nos problemas, entretanto, terá estatísticas muito diferentes umas das outras. Para esse problema temos duas abordagens tradicionais - podemos _normalizar_ a variável, trazendo seu valor mínimo para 0 e máximo para 1 (https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html), ou _padronizar_: mover e escalonar a distribuição para que fique centrada em 0 e tenha desvio padrão 1 (https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html). Para a a _normalização_, subtraímos de todas as amostras o valor mínimo, e dividimos o resultado pelo valor máximo. Para a _padronização_, subtraímos de todas as amostras a média, e dividimos o resultado pelo desvio padrão.

Algoritmos de machine learning que dependem de distâncias entre pontos são muito sensíveis a diferenças de escalas nos dados. Para estes casos, é imprescindível normalizar ou padronizar os dados.

Vamos padronizar nossas amostras anteriores, e normalizar a quantidade amostrada, de forma a ter no eixo $y$ o percentual de amostras com aquele valor:

```python
df['posicao_final'].hist(bins=df['posicao_final'].nunique())
```

```python
# Usaremos a biblioteca sklearn para a padronização
from sklearn.preprocessing import StandardScaler

# Instanciando scaler
scaler = StandardScaler()

# Treinando o estimador
scaler.fit(df['posicao_final'].to_numpy().reshape(-1, 1))

# Aplicando padronização em uma nova coluna
df['posicao_final_padronizada'] = scaler.transform(df['posicao_final'].to_numpy().reshape(-1, 1))

# Criando lista de pontos do eixo x e f(x)
x = np.arange(-5, 5, 0.01)
yg = [gauss(xi, h=max_val/len(df), dp=1) for xi in x]

plt.plot(x, yg)
plt.hist(df['posicao_final_padronizada'], bins=df['posicao_final_padronizada'].nunique(),
         weights=np.ones(len(df)) / len(df))
plt.show()
```
